## Write a solution to find the IDs of the invalid tweets. The tweet is invalid if the number of characters used in the content of the tweet is strictly greater than 15.

Return the result table in any order.

The result format is in the following example.

 

Example 1:

Input: 
Tweets table:
+----------+-----------------------------------+
| tweet_id | content                           |
+----------+-----------------------------------+
| 1        | Let us Code                       |
| 2        | More than fifteen chars are here! |
+----------+-----------------------------------+
Output: 
+----------+
| tweet_id |
+----------+
| 2        |
+----------+

## taking pandas data frame from leet code and than it is converting to spark DataFrame 

```python
import pandas as pd
data = [[1, 'Let us Code'], [2, 'More than fifteen chars are here!']]
tweets = pd.DataFrame(data, columns=['tweet_id', 'content']).astype({'tweet_id':'Int64', 'content':'object'})
```

```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName("Invalid Tweets").getOrCreate()
tweet_df=spark.createDataFrame(tweets)
tweet_df.show()
```

```python
from pyspark.sql.functions import col,length
output_df=tweet_df.filter(length(col("content"))>15).select("tweet_id")
output_df.show()
```
