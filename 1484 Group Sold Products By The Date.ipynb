## 1484. Group Sold Products By The Date
Write a solution to find for each date the number of different products sold and their names.

The sold products names for each date should be sorted lexicographically.

Return the result table ordered by sell_date.

The result format is in the following example.

 

Example 1:

Input: 
Activities table:
+------------+------------+
| sell_date  | product     |
+------------+------------+
| 2020-05-30 | Headphone  |
| 2020-06-01 | Pencil     |
| 2020-06-02 | Mask       |
| 2020-05-30 | Basketball |
| 2020-06-01 | Bible      |
| 2020-06-02 | Mask       |
| 2020-05-30 | T-Shirt    |
+------------+------------+

## creating Spark Session 


```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName("group Sold Product").getOrCreate()
```


## now importing pandas data from leetcode 

```python
import pandas as pd
data = [['2020-05-30', 'Headphone'], ['2020-06-01', 'Pencil'], ['2020-06-02', 'Mask'], ['2020-05-30', 'Basketball'], ['2020-06-01', 'Bible'], ['2020-06-02', 'Mask'], ['2020-05-30', 'T-Shirt']]
activities = pd.DataFrame(data, columns=['sell_date', 'product']).astype({'sell_date':'datetime64[ns]', 'product':'object'})
```

## converting pandas dataFrame to spark data Frame 

```python
df=spark.createDataFrame(activities)
```

## creating temp View of this dataFrame table 

```python
df1=df.createOrReplaceTempView("activity_table")
```

```python
df2 = spark.sql("""
SELECT
    sell_date,
    COUNT(DISTINCT product) AS num_sold,
    CONCAT_WS(
        ',',
        SORT_ARRAY(COLLECT_SET(product))
    ) AS products
FROM activity_table
GROUP BY sell_date
ORDER BY sell_date
""")
```
