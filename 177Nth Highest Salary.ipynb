## 177. Nth Highest Salary
Write a solution to find the nth highest distinct salary from the Employee table. If there are less than n distinct salaries, return null.

The result format is in the following example.

 

Example 1:

Input: 
Employee table:
+----+--------+
| id | salary |
+----+--------+
| 1  | 100    |
| 2  | 200    |
| 3  | 300    |
+----+--------+
n = 2
Output: 
+------------------------+
| getNthHighestSalary(2) |
+------------------------+
| 200                    |
+------------------------+

## Taking pands from leetcode and converting pandas to spark data frame

```python
import pandas as pd
data = [[1, 100], [2, 200], [3, 300]]
employee = pd.DataFrame(data, columns=['Id', 'Salary']).astype({'Id':'Int64', 'Salary':'Int64'})
```

```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName("Nth Salary").getOrCreate()
df=spark.createDataFrame(employee)
df.show()
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import col,dense_rank
winSpec=Window.orderBy(col("Salary").desc())
df1=df.withColumn("sal_rnk",dense_rank().over(winSpec))
df1.show()
```

```python
df2=df1.filter(col("sal_rnk")==2).select("Salary")
df3=df2.withColumn("getNthHighestSalary",col("salary")).select("getNthHighestSalary")
df3.show()
```
