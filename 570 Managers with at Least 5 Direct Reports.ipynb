## Write a solution to find managers with at least five direct reports.

Return the result table in any order.

The result format is in the following example.

 

Example 1:

Input: 
Employee table:
+-----+-------+------------+-----------+
| id  | name  | department | managerId |
+-----+-------+------------+-----------+
| 101 | John  | A          | null      |
| 102 | Dan   | A          | 101       |
| 103 | James | A          | 101       |
| 104 | Amy   | A          | 101       |
| 105 | Anne  | A          | 101       |
| 106 | Ron   | B          | 101       |
+-----+-------+------------+-----------+
Output: 
+------+
| name |
+------+
| John |
+------+

## taking pandas dataFrame from leetcode and then convert them to Spark Session Data Frame

```python
import pandas as pd
data = [[101, 'John', 'A', None], [102, 'Dan', 'A', 101], [103, 'James', 'A', 101], [104, 'Amy', 'A', 101], [105, 'Anne', 'A', 101], [106, 'Ron', 'B', 101]]
employee = pd.DataFrame(data, columns=['id', 'name', 'department', 'managerId']).astype({'id':'Int64', 'name':'object', 'department':'object', 'managerId':'Int64'})
```

```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName("Manager with 5 report").getOrCreate()
emp_df=spark.createDataFrame(employee)
emp_df.show()
```

```python
from pyspark.sql.functions import count,col,cast
mang_df=emp_df.withColumn("managerId",col("managerId").cast("int")).groupBy("managerId").agg(count("*").alias("mang_count")).filter(col("mang_count")>=5).select("managerId")
mang_df.show()
```

```python
from pyspark.sql.functions import col
e=emp_df.alias("e")
m=mang_df.alias("m")
name_df=e.join(m,col("e.id")==col("m.managerId"),"inner").select("name")
name_df.show()
```
